```python
from pydub import AudioSegment
import IPython as inb
from windows_volume_controller import MasterVolumeController

FILE_A = "test_1a.mp3"
FILE_B = "test_1b.mp3"
FILE_B_NORM = "test_1b_normalized.mp3"
```

```python
inb.display.Audio(FILE_A)
```

```python
inb.display.Audio(FILE_B)
```

```python
aud_seg_a = AudioSegment.from_file(FILE_A)
aud_seg_b = AudioSegment.from_file(FILE_B)

dBFS_a = lambda: aud_seg_a.dBFS
dBFS_b = lambda: aud_seg_b.dBFS
```

```python
print(f'before: {FILE_A}: {dBFS_a()}')
print(f'before: {FILE_B}: {dBFS_b()}')
```

```python
change_in_dBFS = dBFS_a() - dBFS_b()
print(f'Change in dBFS: {change_in_dBFS}')
aud_seg_b = aud_seg_b.apply_gain(change_in_dBFS)
print(f'Normalized volume of {FILE_B} according to {FILE_A}')
```

```python
print(f'after: {FILE_A}: {dBFS_a()}')
print(f'after: {FILE_B}: {dBFS_b()}')
```

```python
# Exporting and displaying
aud_seg_b.export(FILE_B_NORM)
inb.display.Audio(FILE_B_NORM)
```

```python
# Setting the volume same as the audio files
volume_controller = MasterVolumeController()
print('before:')
print(f'\tvolume in dB: {volume_controller.get_volume_db()}')
print(f'\tvolume in %: {round(volume_controller.get_volume(), 2) * volume_controller.norm}')
volume_controller.set_volume_db(dBFS_b())
print('after:')
print(f'\tvolume in dB: {volume_controller.get_volume_db()}')
print(f'\tvolume in %: {round(volume_controller.get_volume(), 2) * volume_controller.norm}')
```
